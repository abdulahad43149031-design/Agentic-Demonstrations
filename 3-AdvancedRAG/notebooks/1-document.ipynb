```python
### Text loader
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/python_intro.txt", encoding = "utf-8")
docs = loader.load()

docs[0].page_content
```

```python
from langchain_community.document_loaders import (
    DirectoryLoader,
    TextLoader,
    PyMuPDFLoader,
)

# Load all text files
txt_loader = DirectoryLoader(
    "../data",
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={"encoding": "utf-8"},
)

# We can't do txt and pdfs in a single loader because glob and loader_cls are string, and loader objects respectively and not lists 

# Load all PDF files
pdf_loader = DirectoryLoader(
    "../data",
    glob="**/*.pdf",
    loader_cls=PyMuPDFLoader,
)

pdf_page_docs = pdf_loader.load()
text_docs = txt_loader.load()

docs = pdf_page_docs + text_docs
```
